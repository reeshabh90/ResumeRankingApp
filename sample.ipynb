```python
import json
import os
from typing import List
import logging
from scripts.similarity import get_similarity_score, find_path, read_config
from scripts.utils import get_filenames_from_dir
```

```python
def read_json(filename):
    with open(filename) as f:
        data = json.load(f)
    return data
```

```python
selected_file = read_json("Data/Processed/Resumes/" + "Resume-alfred_pennyworth_pm.pdf4734d380-8715-4008-b2cf-d7097c38bd79.json")
selected_jd = read_json(
    "Data/Processed/JobDescription/" + "JobDescription-job_desc_front_end_engineer.pdfebf13f7a-be90-478b-baab-b64ba2807fd4.json")
```

```python
resume_dict = read_json(
        READ_RESUME_FROM + "/Resume-alfred_pennyworth_pm.pdf4734d380-8715-4008-b2cf-d7097c38bd79.json")
job_dict = read_json(
        READ_JOB_DESCRIPTION_FROM + "/JobDescription-job_desc_front_end_engineer.pdfebf13f7a-be90-478b-baab-b64ba2807fd4.json")
resume_keywords = resume_dict["extracted_keywords"]
job_description_keywords = job_dict["extracted_keywords"]
resume_keywords
```

```python
resume_string = ' '.join(resume_keywords)
jd_string = ' '.join(job_description_keywords)
resume_string
```
